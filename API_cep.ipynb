### Via CEP
- endpoint, link do tipo GET
- https://viacep.com.br/

- pip install requests

#### Consulta de informações de um CEP

```python
import requests

cep = "20.090-002"

cep = cep.replace("-", "").replace(".", "").replace(" ", "") # faz um tratamento no CEP

if len(cep) == 8: # Se o tamanho do CEP for igual a 8 faz a requisicao, caso contrário informa CEP inválido.
    link = f'https://viacep.com.br/ws/{cep}/json/'

    requisicao = requests.get(link)

    print(requisicao)

    dic_requisicao = requisicao.json()

    uf = dic_requisicao['uf']
    cidade = dic_requisicao['localidade']
    bairro = dic_requisicao['bairro']
    print(dic_requisicao)
else:
    print("CEP Inválido")
```

#### Busca de CEP a partir de endereço

```python
uf = "RJ"
cidade = "Rio de Janeiro"
endereco = "Rio Branco"

link = f'https://viacep.com.br/ws/{uf}/{cidade}/{endereco}/json/'

requisicao = requests.get(link)
print(requisicao)

dic_requisicao = requisicao.json()
print(dic_requisicao)
```
